### Unveiling Accident Trends with Heatmaps

The goal is simple: to spot trends and patterns in accident data using heatmaps. Heatmaps offer a unique perspective, quickly revealing areas of high and low accident concentration. This visualization technique will help identify potential safety concerns and hot spots, providing valuable insights for preventive measures.

```python
import folium
from folium.plugins import HeatMap
map_center = [42.3876, -71.0995]
m = folium.Map(location=map_center, zoom_start=14)


locations = df[['Latitude', 'Longitude']].to_numpy()


HeatMap(locations, radius=20, blur=15).add_to(m)

# Display the map
m
```

### Observation:

The analysis reveals a uniform distribution of crashes across Smoreville City, indicating no apparent concentration or pattern in terms of location. This suggests that accident risks are evenly spread throughout the city, without any specific areas of concern. However, further investigations and data analysis are warranted to uncover potential underlying factors contributing to these accidents.

### Unveiling Crash Trends by Time of Day

This analysis aims to uncover patterns in accident data by examining the time of day. By understanding when crashes occur, we can identify high-risk periods and adjust safety measures accordingly. Visualizing this data will provide insights into peak accident times, helping authorities optimize resources and target interventions for maximum impact.

```python
def convert_to_hourly(datetime_str):
  datetime = pd.to_datetime(datetime_str)
  hour = datetime.hour + round((datetime.minute / 60), 0)  # Round to nearest hour
  return hour

df['Hour'] = df['Date and Time of Crash'].apply(convert_to_hourly)
```

```python
crushes_by_hour = df['Hour'].value_counts()

sns.set_theme(style="darkgrid")

plt.bar(crushes_by_hour.index, crushes_by_hour.values)
plt.xlabel("Hours of the Day")
plt.ylabel("Number of Crashes")
plt.title("Crash Counts by Hour")

plt.show()
```

### Conclusion 

Analysis confirms expected patterns: crashes decrease at night, increase during the day, and remain steady. No specific hour stands out, indicating consistent daytime risk.
